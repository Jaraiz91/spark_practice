### As all columns are strings we must convert them to their different types of data

```
from pyspark.sql.types import StructField, StructType, DoubleType, IntegerType, TimestampType, StringType, FloatType

```

```
data_schema = [StructField('Date', TimestampType(), True), StructField('Open', DoubleType(), True), StructField('High', DoubleType(), True), StructField('Low', DoubleType(), True), StructField('Close', DoubleType(), True),
              StructField('volume', IntegerType(), True), StructField('Adj Close', DoubleType(), True)]
```

```
df_structure = StructType(fields = data_schema)
df = spark.read.format("csv").option("header","true").load("dbfs:/FileStore/shared_uploads/dataraiz91@gmail.com/walmart_stock.csv", schema= df_structure)
df.printSchema()
```

```
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('walmart').getOrCreate()
```

```
describe_schema = [StructField('summary', StringType(), True), StructField('Open', FloatType(), True), StructField('High', FloatType(), True),
                  StructField('Low', FloatType(), True), StructField('Close', FloatType(), True), StructField('volume', FloatType(), True), StructField('Adj Close', FloatType(), True)]
describe_structure = StructType(fields= describe_schema)
describe_df = spark.createDataFrame(describe_data, schema=describe_structure)
```

```

describe_df.select('summary',format_number('Open',2).alias('Open'), 
                  format_number('Close', 2).alias('Close'), format_number('High', 2).alias('High'),
                  format_number('Low', 2).alias('Low'), format_number('volume', 2).alias('volume')).show()
```

#### What is the average Close for each Calendar Month?
#### In other words, across all the years, what is the average Close price for Jan,Feb, Mar, etc... Your result will have a value for each of these months.

```
from pyspark.sql.functions import month
```

```
df_month = df.withColumn('Month', month(df['Date']))
df_month = df_month.groupBy('Month').mean()
df_month = df_month.orderBy('Month')
df_month.select('Month', 'avg(Close)').show()
```
